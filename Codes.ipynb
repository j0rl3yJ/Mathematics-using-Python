## Use Sympy to understand law of exponents

```python
"""
If an expression cannot be solved using simplify method of sym module, then the answer shown will be
the same as the original expression.
"""
import sympy as sym
from IPython.display import display, Math

# Create 3 variables
x, y, z = sym.symbols('x, y, z')

# Create an expression
expr1 = (x**y) * (x**z)
expr2 = (x**y) * (x**-z)
expr3 = (x**y) * (y**z)

# Display the expression
display(Math('%s' %sym.latex(expr1)))
display(Math('%s' %sym.latex(expr2)))
display(Math('%s' %sym.latex(expr3)))

# Simplify the equation
display(sym.simplify(expr1))
display(sym.simplify(expr2))
display(sym.simplify(expr3))

print("Better way of printing the equation in an elegant way:")
display(Math('%s = %s' %(sym.latex(expr1), sym.latex(sym.simplify(expr1)))))

"""
I can also use the Eq() function of the sym module. 
This equates both the LHS and RHS and creates a boolean value of True or False.
The default for the second position argument is zero.
"""
#This checks whether the difference of the two expressions is zero and evaluates to a boolean value.
sym.Eq(sym.expand(expr1 - sym.simplify(expr1)))
```

```python
# Import the libraries
import numpy as np
import sympy as sym
from IPython.display import display, Math

# Create a homogenous equation:
"""
So basically I am creating an equation which is set equal to zero, is implicit. 2X + 4 = 9.
"""
X = sym.symbols('X')
expression = 2*X + 4 - 9
# Solve for X using the solve module of the sympy library
solution = sym.solve(expression)[0]
display(Math('\\text{The solution to } 2X + 4 - 9\\text{ is X = }%g'%solution))

"""
Suppose if we have multiple solutions, this is why the output of the solve function is a list.
"""
# Create an expression
expression = X**2 - 4
# Get the solution ready
solution = sym.solve(expression)
# Iterate over the list of solutions to get the output printed using a for loop.
display(Math('\\text{Solution to } X^2 - 4 \\text{ is :}'))
for i in range(0, len(solution)):
    print("\tSolution #%g is : %g" %(i+1, solution[i]))
    
"""
Now suppose I have an equation of multiple variables. After solving it, I will get a dictionary of 
solutions.
To make it easier to read, either use a loop or give the variable to be solved for in the solve function.
"""
Y = sym.symbols('Y')
expression = X/4 - X*Y + 5
#Solution for X:
display(Math('\\text{Solution is : } %s' %(sym.latex(sym.solve(expression,X)))))
# Solution for Y:
display(Math('\\text{Solution is : } %s' %(sym.latex(sym.solve(expression,Y)))))

# Exercise 1:
"""
Implement this equation
Simplify
Solve for q
"""
q = sym.symbols('q')
display(Math('3q + \\frac{4}{q} + 3 = 5q + \\frac{1}{q} + 1'))
equation = (3*q) + (4/q) + 3 - (5*q) - (1/q) - 1
equation = sym.simplify(equation)
solution = sym.solve(equation, q)
display(Math('%s' %sym.latex(solution)))

# Exercise 2:
"""
Implement the equation in simpy
apply the functions simplify and cancel
read help text to understand the difference in outputs.
"""
display(Math('2q + 3q^2 - \\frac{5}{q} - \\frac{4}{q^3}'))
equation = (2*q) + (3*q**2) - (5/q) - (4/q**3)
display(Math('%s' %sym.latex(sym.cancel(equation))))
display(Math('%s' %sym.latex(sym.simplify(equation))))

# Exercise 3:
"""
Implement the equation
simplify
susbtitute and check for q.
"""
import numpy as np
display(Math('\\text{Exercise 3 equation is : } \\frac{\\sqrt3 + \\sqrt15q}{\\sqrt2 + \\sqrt10q}'))
equation = (np.sqrt(3) + (np.sqrt(15)*q))/(np.sqrt(2) + (np.sqrt(10)*q))
solution = sym.simplify(equation)
#display the solution:
display(Math('\\text{Solution to the equation is : } %s' %sym.latex(solution)))
#substitute for q 
solution.subs(q, 0).evalf()
```

## Expanding terms:

```python
"""
use sym.expand() to expand the terms and then embed the silution into latex, to get a neat looking style.
sym.expand() will also work with multiple variables.
"""
import sympy as sym
from IPython.display import display, Math
x, y = sym.symbols('x, y')
term1 = (4*x + 5)
term2 = x
term3 = (x - 7)
term4 = 2*y**2 - 5**x/x
display(Math('\\text{Term 1 is : } %s' %sym.latex(term1)))
display(Math('\\text{Term 2 is : } %s' %term2))
display(Math('\\text{Term 3 is : } %s' %term3))
display(Math('%s \\times %s \\text{ is : } %s'%(sym.latex(term1), term2, 
                                                sym.latex(sym.expand(term1 * term2)))))
display(Math('%s \\times %s \\text{ is : } %s'%(sym.latex(term1), term3, 
                                                sym.latex(sym.expand(term1 * term3)))))
display(Math('%s \\times %s \\text{ is : } %s'%(term2, sym.latex(term4), 
                                                sym.latex(sym.expand(term2 * term4)))))

# Exercise: 
"""
substitute values ranging from 0 through 2 in the equation
When you have to substitute values for multiple variables in a solution, use dictionaries for the multiple
    variables.
"""
# Declare the terms:
term1 = 4 + x
term2 = 2 - y

# Display the terms and the boundary values:
display(Math('f(x, y) = (%s)\\times(%s)' %(term1, term2)))
display(Math('x,y \\in \\text{{0, 1, 2}}'))

# Make solution:
solution = sym.expand(term1*term2)
display(Math(sym.latex(solution)))

# Substitute different values 
for i in range(0, 3):
    for j in range(0, 3):
        display(Math('\\text{When x = } %g \\text{ and y = } %g \\text{, f(x, y) = } %g' 
                     %(i, j, solution.subs({x:i, y:j}))))
```

## Creating and accessing matrices using Numpy

```python
"""
You can also populate the zero matrix, and use it further.
"""
import numpy as np
import sympy as sym
from IPython.display import Math, display

# Create a 2D matrix
A = np.array([[1,2], [3,4]])
display(Math(sym.latex(sym.sympify(A))))

# Create a matrix of zeros
matrix = np.zeros([4,6])
matrix[0,1] = 2
display(Math(sym.latex(sym.sympify(matrix))))

# Exercise:
"""
substitute values ranging from 0 through 2 in the equation
When you have to substitute values for multiple variables in a solution, use dictionaries for the multiple
    variables.
"""
# Declare the terms and matrix:
x, y = sym.symbols('x,y')
term1 = 4 + x
term2 = 2 - y
matrix = np.zeros([3,3])

# Display the terms and the boundary values:
display(Math('f(x, y) = (%s)\\times(%s)' %(term1, term2)))
display(Math('x,y \\in \\text{{0, 1, 2}}'))
display(Math(sym.latex(sym.sympify(matrix))))

# Make solution:
solution = sym.expand(term1*term2)
display(Math(sym.latex(solution)))

# Substitute different values 
for i in range(0, 3):
    for j in range(0, 3):
        matrix[i,j] = solution.subs({x:i,y:j})
        
# Display the final matrix:
display(Math(sym.latex(sym.sympify(matrix))))

# Exercise:
"""
Create a multiplication table:
"""
# Create a 10x10 matrix:
multiplication_table = np.zeros([10,10],dtype=int)

# Populate the matrix:
for row in range(1,11):
    for column in range(1,11):
        multiplication_table[row-1,column-1] = row*column
        
# Display the matrix:
display(Math(sym.latex(sym.sympify(multiplication_table))))
```

## Associative, Commutative and Distributive property

```python
"""
Associative rule:
ax(bxc) = (axb)xc

Commutative rule:
axb = bxa

Distribute rule:
ax(b+c) = ab + ac
"""
import numpy as np
import sympy as sym
from IPython.display import display,Math

# Exercise:
"""
1. Implement these expressions in Python
2. Display and simplify f1xf2
3. Show that the commutative property holds.
"""
# Declare the variables:
w,x,y,z = sym.symbols('w,x,y,z')

# Declare the two expressions and display:
x = w*(4-w) + ((1/w**2)*(1+w))
f1 = x * (y + z)
f2 = (3/x) + x**2
display(Math('x = %s'%sym.latex(x)))
display(Math('f_1 = %s'%sym.latex(f1)))
display(Math('f_2 = %s'%sym.latex(f2)))

# Simplify f1xf2:
solution = sym.simplify(f1 * f2)
display(Math('f_1 \\times f_2 = %s' %sym.latex(solution)))

# To show that commutative property holds:
print("\nDoes the Commutative property hold : %s" %(sym.Eq(sym.simplify(f1*f2) - sym.simplify(f2*f1), 0)))
```
