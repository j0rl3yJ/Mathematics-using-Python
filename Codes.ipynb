## Use Sympy to understand law of exponents

```python
"""
If an expression cannot be solved using simplify method of sym module, then the answer shown will be
the same as the original expression.
"""
import sympy as sym
from IPython.display import display, Math

# Create 3 variables
x, y, z = sym.symbols('x, y, z')

# Create an expression
expr1 = (x**y) * (x**z)
expr2 = (x**y) * (x**-z)
expr3 = (x**y) * (y**z)

# Display the expression
display(Math('%s' %sym.latex(expr1)))
display(Math('%s' %sym.latex(expr2)))
display(Math('%s' %sym.latex(expr3)))

# Simplify the equation
display(sym.simplify(expr1))
display(sym.simplify(expr2))
display(sym.simplify(expr3))

print("Better way of printing the equation in an elegant way:")
display(Math('%s = %s' %(sym.latex(expr1), sym.latex(sym.simplify(expr1)))))

"""
I can also use the Eq() function of the sym module. 
This equates both the LHS and RHS and creates a boolean value of True or False.
The default for the second position argument is zero.
"""
#This checks whether the difference of the two expressions is zero and evaluates to a boolean value.
sym.Eq(sym.expand(expr1 - sym.simplify(expr1)))
```

## Associative, Commutative and Distributive property

```python
"""
Associative rule:
ax(bxc) = (axb)xc

Commutative rule:
axb = bxa

Distribute rule:
ax(b+c) = ab + ac
"""
import numpy as np
import sympy as sym
from IPython.display import display,Math

# Exercise:
"""
1. Implement these expressions in Python
2. Display and simplify f1xf2
3. Show that the commutative property holds.
"""
# Declare the variables:
w,x,y,z = sym.symbols('w,x,y,z')

# Declare the two expressions and display:
x = w*(4-w) + ((1/w**2)*(1+w))
f1 = x * (y + z)
f2 = (3/x) + x**2
display(Math('x = %s'%sym.latex(x)))
display(Math('f_1 = %s'%sym.latex(f1)))
display(Math('f_2 = %s'%sym.latex(f2)))

# Simplify f1xf2:
solution = sym.simplify(f1 * f2)
display(Math('f_1 \\times f_2 = %s' %sym.latex(solution)))

# To show that commutative property holds:
print("\nDoes the Commutative property hold : %s" %(sym.Eq(sym.simplify(f1*f2) - sym.simplify(f2*f1), 0)))
```

## Greatest Common Denominator:

```python
"""
It is the greatest divisor that divides both the numbers cleanly without any remainders.
One of the application of GCD is reducing fractions.
Can only do gcd of integers using math.gcd
"""
import math
import numpy as np
import sympy as sym
from IPython.display import display, Math

# print the gcd of two numbers:
print(math.gcd(6, 15))

# Exercise:
"""
Illustrate that gcd of two numbers = gcdxthe two numbers
"""
a, b, c = sym.symbols('a, b, c')
display(Math('%s = %s' %((sym.gcd(a*b, a*c)), (a*sym.gcd(b, c)))))
a = 10
b = 15
factor = math.gcd(a, b)
display(Math('\\text{Is } %g \\text{ the gcd of } %g \\text{ and } %g \\text{ : } %s' 
             %(factor, a, b, sym.Eq(factor - factor*(math.gcd(int(a/factor), int(b/factor)))))))

"""
Create a matrix that contains 10 rows and 15 columns, where each element is equal to 99.

Populate the matrix such that ith rrow and jth column contains gcd(i+1, j+1).
"""
# Create a 2D matrix initialized to 99:
matrix = np.zeros([10,15], dtype=int) + 99

# Populate the matrix such that ith row and jth column is gcd(i+1, j+1)
for row in range(0,10):
    for column in range(0,15):
        matrix[row,column] = math.gcd(row+1, column+1)

# Display the final matrix:
display(Math('%s'%sym.latex(sym.sympify(matrix))))
```

## Prime factorization:

```python
"""
Loop through the interval [2,50] and print if the number is prime or composite.
If composite, list the prime numbers.

to find the prime factors, use factorint function from sympy module.
"""
import sympy as sym

# Create a list of numbers:
integer_list = [i for i in range(2, 51)]

for number in integer_list:
    factor_dictionary = sym.factorint(number)
    factor_keys = list(factor_dictionary.keys())
    if factor_dictionary[factor_keys[0]] == 1 and len(factor_keys) == 1:
            print("%g is a prime number." %number)
    else:
        print("%g is a composite number with prime factor(s) of %s" 
              %(number, list(factor_dictionary.keys())))
```

## Adding Polynomials

```python
import sympy as sym
from IPython.display import display, Math

# Declare the polynomials:
p1 = sym.poly(2*x**3 + x**2 - x)
p2 = sym.poly(x**3 - x**4 - 4*x**2)

# Display the polynomial:
display(Math(sym.latex(p1)))
display(Math(sym.latex(p2)))

# Add the polynomials:
solution = p1 + p2
display(Math('(%s) + (%s) = %s' %(sym.latex(p1), sym.latex(p2), sym.latex(solution))))

# Evaluate the polynomial to some value 0:
print("Polynomial evaluated to 0 : %g" %solution.eval(0))

# Display the coefficients of the solution:
print("Coeffecients of the solution : %s" %solution.coeffs())

# Exercise:
"""
Implement the polynomials as list.
For each polynomial, if the degree is even : sum the coeffecients
                     if the degree is odd : count the number of coeffecients
Hint: use poly.degree() -> to find the highest degree
"""
# Declare the polynomials:
p1 = sym.poly(x**2 + 2*x)
p2 = sym.poly(-x**3 + 4*x)
p3 = sym.poly(x**5 - x**4 + (x/4) +4)

# Put the polynomials in a list and display:
polynomial_list = [p1, p2, p3]
for polynomial in polynomial_list:
    display(Math(sym.latex(polynomial)))
    
    #If the degree is odd, count the number of coeffecients:
    if polynomial.degree() % 2 != 0:
        print("Number of coeffecients of the polynomial is : %g" %len(polynomial.coeffs()))
    
    # Else if degree is even, sum the coeffecients:
    else:
        print("The sum of the coeffecients of the polynomial is : %g" %sum(polynomial.coeffs()))
```

## Multiplying Polynomials

```python
"""
To solve two polynomials, use the sym.expand() method.
"""
import sympy as sym
from IPython.display import display, Math

# Declare the two polynomials:
x = sym.symbols('x')
p1 = 4*x**2 - 2*x
p2 = x**3 - 1

# Multiply the two polynomials:
display(Math('(%s) \\times (%s) = %s' %(sym.latex(p1), sym.latex(p2), sym.latex(sym.expand(p1 * p2)))))

# Exercise:
#Implement the polynomials:
x, y = sym.symbols('x, y')
f = 4*x**4 - 3*x**2 + x*y**2 - 9*y**3
g = -x**3 + 6*(x**2)*y + .8*y**3
display(Math('\\text{f(x,y) = } %s' %sym.latex(f)))
display(Math('\\text{g(x,y) = } %s' %sym.latex(g)))

# Multiply the two polynomials and then substitute x = 5, y = -2
print("Multiplying the two polynomials and then substituting x = 5, y = -2....\nAnswer : ")
solution = sym.expand(f * g)
display(Math('%s' %sym.latex(solution)))
display(Math('%g'%solution.subs({x : 5, y : -2})))

# Substitute x = 5 and y = -2 into each polynomial, then multiply:
print("Substituting x = 5 and y = -2 into each polynomial, then multiplyting...\nAnswer : ")
display(Math('%g \\times %g = %g' %(f.subs({x : 5, y : -2}), g.subs({x : 5, y : -2}), 
                                    f.subs({x : 5, y : -2}) * g.subs({x : 5, y : -2}))))
```

```python
"""
To divide polynomials and get the quotient, use the sym.simplify() method.
"""
import sympy as sym
from IPython.display import display, Math

# Declare and initialize the polynomials:
x = sym.symbols('x')
p1 = 4*x**5 - x
p2 = 2*x**3 - x

# Solve and display the result:
solution = sym.simplify(p1/p2)
display(Math('\\frac{%s}{%s} = %s' %(sym.latex(p1), sym.latex(p2), sym.latex(solution))))

# Exercise:
"""
For the following polynomial division, find the value for y such that denominator = 1.
"""
import sympy as sym
from IPython.display import display, Math

# declare and initialize the polynomial fraction:
x, y = sym.symbols('x, y')
p1 = (x**6 + 2*x**4 + 6*x - y)
p2 = (x**3 + 3)
expression = (p1/p2)
display(Math('\\frac{%s}{%s} \\text{, y} \\in \\text{{5,6,....,15}}' %(sym.latex(p1), sym.latex(p2))))

# Solve iteratively, 5 through 15 to find the denominator = 1:
for i in range(5, 16):
    p1_substituted = p1.subs(y, i)
    solution = sym.simplify(p1_substituted/p2)
    if (sym.fraction(solution)[1] == 1):
        print("For y = %g, the denominator becomes zero." %i)
        display(Math('%s' %sym.latex(solution)))
    else:
        continue
```

## Factoring Polynomials:

```python
"""
To do factoring of polynomials, use the sym.factor() method.
"""
# Exercise:
"""
For the following polynomials, if it can be factored, factor the polynomial and then display it.
Else, display not factorable!.
"""
import sympy as sym
from IPython.display import display, Math

# Declare and initialize the polynomials.
x, y = sym.symbols('x, y')
p1 = x**2 + 4*x + 3
p2 = 2*y**2 - 1
p3 = 3*y**2 + 12*y

# Create a polynomial list and iterate over the list to check if it is factorable:
polynomial_list = [p1, p2, p3]
for polynomial in polynomial_list:
    solution = sym.factor(polynomial)
    if('(' in str(solution)):
        display(Math("%s \\quad \\Rightarrow \\quad %s." %(sym.latex(polynomial), sym.latex(solution))))
    else:
        display(Math("%s \\quad \\Rightarrow \\quad \\text{ is not Factorable!}" %sym.latex(polynomial)))
```

## Plotting cordinates on a plane:

```python
"""
Plotting single cordinate and multiple cordinates:
"""
import matplotlib.pyplot as plt

# Declare and initialize the cordinates:
x, y = 5, 3

# Plot it:
plt.plot(x, y, 'bo')
# Make the plot square:
plt.axis('square')
# Add the axis limits. First two are x axis limits. Last two are y axis limits:
plt.axis([0,6,0,6])
# Add grid:
plt.grid()
# Show the plot:
plt.show()

"""
Plotting multiple points on the cordinate system:
"""
x = [1,2,3,4,5,6,7]
y = [2,6,8,9,3,4,6]

# use for loop to traverse through the length of the list:
for i in range(0, len(x)):
    plt.plot(x[i], y[i], 'o', label = '(%s, %s)' %(x[i], y[i]))
    
plt.axis('square')
plt.axis([0,10,0,10])
plt.legend()
plt.grid()

plt.show()

"""
Suppose I want to change the limit bounds of the axis:
"""
plt.plot(3,4,'bo', label = '(%s,%s)'%(3,4))
# get the info about the axis. gca = get current axis:
axis = plt.gca()
# get info about the y axis:
y_limits = axis.get_ylim()
print(y_limits)
# change y limits:
axis.set_ylim(3, 5)
# activate the legends:
plt.legend()
plt.show()

# Exercise:
"""
Plot the following equation y = x**2 - 3*x in the interval x in [-10, 11]
"""
import sympy as sym
from IPython.display import display, Math
import matplotlib.pyplot as plt
sym.init_printing()

# Display the equation:
x = sym.symbols('x')
equation = x**2 - 3*x
display(Math('\\text{Equation to plot : } %s' %sym.latex(equation)))

# Plot the values for x and y:
for i in range(-10, 12):
    plt.plot(i, equation.subs(x,i), 'ko')
plt.xlabel('x')
plt.ylabel('f(x) = x**2 - 3*x')
plt.show()
```

## Graphing lines by slope-intercept form

```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt

# declare the x-intercept and slope:
b = [-2,0.75]
m = [0.7,-1.25]
x = [-5,5]
for i in range(len(x)):
    plt.plot(x, m[i]*np.array(x)+b[i], label = 'y = %sx + %s'%(m[i], b[i]))
    
plt.grid()
plt.axis('square')
axis = plt.gca()
plt.plot(axis.get_xlim(), [0,0], 'k--')
plt.plot([0,0], axis.get_ylim(), 'k--')
plt.xlim(x)
plt.ylim(x)
plt.legend()
plt.show()
```
